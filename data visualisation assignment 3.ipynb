```python
for col in ['Name','City']:
    mask=df[col].str.len() !=df[col].str.strip().str.len()
    if mask.any():
        print(df.loc[mask,col])
        df[['Name','City']]=df[['Name','City']].apply(lambda x: x.str.strip())
        print ("\nSTEP 1_After trimming whitespace:\n ")
        print(df.head(8))
```

```python
#Step 2 - Fix inconsistent text case in City column
#Problem: Cities are written with different cases (Delhi, delhi, DELHI, etc.).
#Goal: Standardize them (for example â†’ Title Case: Delhi).
# STEP 2 - Standardize City column (title case)
df['City'] = df[ 'City'].str.title()
print("STEP 2 - After standardizing City column: \n" )
print(df[ 'City'].unique())
# check unique city names
```

```python
# Non-numeric entries (like "Twenty")
# Should be integer numbers
# Code for Cleaning Age
import numpy as np
# STEP 3 - Clean Age column
# 1. Replace word-based numbers manually (if limited cases)
df[ 'Age'] = df[ 'Age' ].replace({
'Twenty': 20,
'Thirty': 30,
'Forty': 40
})
# 2. Convert Age to numeric, forcing errors to NaN
df[ 'Age'] = pd.to_numeric(df['Age'], errors='coerce')
# 3. Optional: Fill missing/invalid ages with median
median_age = df[ 'Age' ].median()
df['Age']=df['Age'].fillna(median_age)
print("STEP 3  -after cleaning age column;\n")
print(df[['ID','Name','Age']].head(10))
```

```python
# Inconsistent formats (48k, 50,000, extra spaces)
# Needs to be numeric for analysisCode for Cleaning Salary
# STEP 4 - Clean Salary column
# 1. Remove commas and spaces
df[ 'Salary'] = df['Salary'].astype(str).str.replace(",", " ").str.strip()
# 2. Handle values like "40k" -> 40000
df[ 'Salary'] = df['Salary'].str. replace("k", "000", case=False)
#. 3. Convert to numeric (invalid values -> NaN)
df[ 'Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
#Optional: Fill missing salaries with median
median_salary = df['Salary' ].median()
df[ 'Salary'] = df[ 'Salary'].fillna(median_salary)
print("STEP 4 - After cleaning Salary column:\n")
print(df[['ID', 'Name', 'Salary' ]].head(10))
```

```python
# Let's wrap it up with Step 5 - Final consistency check.
# Code for Step 5
# STEP 5 - Final consistency check
print("STEP 5 - Shape of dataset:", df.shape)
print("\nColumn data types:")
print(df.dtypes)
print("\nMissing values after cleaning:")
print(df.isna().sum())
print("InSummary statistics:")
print(df.describe(include='all'))
print("\nFinal snapshot of cleaned dataset:")
print(df.head(10))
```

```python
# Let's make it a separate Step 6 - Handle missing values in categorical columns.
# STEP 6 - Handle missing City values
# Option 1: Fill with "Unknown"
df['City'] = df['City'].fillna ("Unknown")
# OR Option 2: Fill with most frequent city (mode)
#mode_city = df[ 'City'].mode() [0]
# df["City' ] = df['City'].fillna(mode_city)
print("STEP. 6 - After handling missing City values: \n")
print(df[ 'City' ].unique())
print("\nMissing values check:\n", df.isna().sum())
```

```python
# Let's finalize by exporting the cleared dataset.
# STEP 7 - Export cleaned dataset to CSV
output_file = "Cleaned.csv"
df.to_csv(output_file, index=False)
print(f"STEP 7 - Cleaned dataset exported successfully as (output_file)")
```
